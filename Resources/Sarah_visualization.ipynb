```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
# CSV into dataframe
csv = "../output/result2.csv"
result_df = pd.read_csv(csv)
result_df.head()
```

```python
# Plot the Bad Word Count vs Rating and save figure
plt.scatter(x=result_df['IMDB'], y=result_df['total_bad_words'])
plt.title('Bad Word Count Vs. Rating')
plt.ylabel('Bad Word Count')
plt.xlabel('IMDB Rating')
plt.grid()
plt.savefig('../Figures/Count_v_Rating.png')
plt.show()
```

```python
# Group average Bad Word Count by year
df_group = result_df.groupby(by='Year').mean()
df_group.head()
```

```python
# Reduce dataframe to year and average bad word count
df_group_bad = df_group['total_bad_words']
```

```python
# Plot average bad word count by year and save figure
df_group_bad.plot()
plt.title('Mean Bad Word Count by Year')
plt.ylabel('Bad Word Count')
plt.grid()
plt.savefig('../Figures/Count_by_Year.png')
plt.show()
```
