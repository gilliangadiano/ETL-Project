```python
import requests
```

```python
# store the URL for the list of movies
url = 'https://www.imsdb.com/all%20scripts/'
```

```python
# Make the request and check object type
r = requests.get(url)
type(r)
```

```python
# Extract HTML from Response object and print
html = r.text
print(html)
```

```python
# Import BeautifulSoup from bs4
from bs4 import BeautifulSoup

# Create a BeautifulSoup object from the HTML
soup = BeautifulSoup(html, "html5lib")
type(soup)
```

```python
# Pull a list of avaiable scripts and create a link dictionary
href_list = soup.findAll('a', class_='')
movie_links = [{}]
title_list = []
```

```python
# Add movie titles and URLs to the movie_links dictionary
for href in href_list:
    links = href.get('href')
    title = href.get('title')
    title_list.append(title)
    movie_links.append({title:links})
```
